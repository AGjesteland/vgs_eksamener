```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
FONTSIZE = 15
```

## Del 1 - oppgave 6

```python
x = np.linspace(-6, 6, num = 2000)
f = 6 / (1 + np.exp(-x))
f_deriv = 6 * np.exp(-x) / (1 + np.exp(-x))**2


fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title('Skisse av $f(x) = 6 \, (1 + e^{-x})^{-1}$', fontsize = FONTSIZE)
plt.plot(x, f, linewidth=2, label = '$f(x)$')
plt.plot(x, f_deriv, '--', linewidth=1.5, label = "$f'(x)$")
plt.grid(True, ls='--')


#plt.xlabel('$x$', fontsize = FONTSIZE)
#plt.ylabel('$y$', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xlim([-5, 5])
plt.ylim([-1, 7])
plt.tight_layout()
plt.savefig('../figs/oppg_6d.pdf')
plt.show()
```

## Oppgave 3

```python
x = np.linspace(-6, 6, num = 2000)
P = x**3 - 3*x**2 - 13*x + 15
P_deriv = 3*x**2 - 6*x**1 - 13

fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title('Skisse av $P(x) = x^3 - 3x^2 - 13x + 15$', fontsize = FONTSIZE)
plt.plot(x, P, linewidth=2, label = '$P(x)$')
plt.plot(x, P_deriv, '--', linewidth=1.5, label = "$P'(x)$")
plt.grid(True, ls='--')


#plt.xlabel('$x$', fontsize = FONTSIZE)
#plt.ylabel('$y$', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xlim([-4, 6])
plt.ylim([-30, 30])
plt.tight_layout()
plt.savefig('../figs/oppg_3.pdf')
plt.show()
```
